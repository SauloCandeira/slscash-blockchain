```python
#Ethereum Smart Contracts With Python - Web3.py Nº: 3
from web3 import Web3

ganache_url = "HTTP://127.0.0.1:7545"
web3 = Web3(Web3.HTTPProvider(ganache_url))
print(web3.isConnected())
print(web3.eth.blockNumber)

account_1 = '0x3c9f39b5b2beA0CA4E447C998e315F7a1fFC6150'
account_2 ='0x2FE327d33382fa801A2c6174BB142929bc9663bb'

private_key='307311b47f659f4bb75bdd74465752218f8815ef74d2af1b231c9c678ce0dda0'

nonce = web3.eth.getTransactionCount(account_1)

tx = {
    'nonce': nonce,
    'to' : account_2,
    'value': web3.toWei(1,'ether'),
    'gas': 2000000,
    'gasPrice': web3.toWei('50','gwei') 
}

signed_tx = web3.eth.account.signTransaction(tx, private_key)
tx_hash = web3.eth.sendRawTransaction(signed_tx.rawTransaction)
print(web3.toHex(tx_hash))
```
